# Create sql file using python

```python
import sqlite3
def gen_insert_query(table_name:str, a_dict:dict) -> str:
    # https://sql.sh/cours/insert-into
    # INSERT INTO table-name (column-names)  VALUES (values) 
 
    insert_query_tpl = "INSERT INTO {} {} VALUES {} "
    keys, values = zip(*a_dict.items())
     
    compiled_query = insert_query_tpl.format(table_name, keys, values)
     
    return compiled_query

#Tentative de batch, mais impossible de commit plus de 50 fois (cf. transactions.).

# def gen_insert_file(filename,tablename,df): 
#     dict_row={}
#     df_c=df.dropna().copy()
#     i=0
#     conn = sqlite3.connect(filename)
#     for k in range(0,len(df_c)): 
        
#         c = conn.cursor()
#         for i in range(0,len(df_c.loc[k].values)):
#             dict_row[df_c.columns.values[i]]=df_c.loc[k].values[i]
#         query=gen_insert_query(tablename,dict_row)
#         print(query)
#         c.execute(query)
#         conn.commit()
#     conn.close()
```

```python
def gen_insert_file(tablename:str, df):
    df_c=df.copy()
    values = [tuple(x) for x in df_c.to_numpy()]
    filename=tablename+"_insert.sql"
    f= open(filename,"w+")
    f.write("BEGIN;\n")
    
    for i in range(0,len(df_c)):
        if 'nan' in format(values[i]):
            va=format(values[i]).replace("nan", "NULL")
            f.write("INSERT INTO {} {} VALUES {};\n".format(tablename, tuple(df_c.columns.values), va))
        else:
            f.write("INSERT INTO {} {} VALUES {};\n".format(tablename, tuple(df_c.columns.values), values[i]))
    f.write("END TRANSACTION;")
    f.close
    
%lprun -f gen_insert_file gen_insert_file('trips',df_trips)
```

```python
gen_insert_file('trips',df_trips)
gen_insert_file('stops',df_stops)
gen_insert_file('stop_times',df_stop_times)
gen_insert_file('agency',df_agency)
gen_insert_file('routes',df_routes)
```

# KDtree: finds nearest neighbours and time (work in progress)

```python
from scipy import spatial

import pysal
from pysal.cg.kdtree import KDTree

```

```python
X = (45.186308,5.727858)
x=list(df_stops.stop_lat.values)
y=list(df_stops.stop_lon.values)
stops_coordinates=[(x[k],y[k]) for k in range(0,len(x))]
```

```python
tree=KDTree(stops_coordinates,distance_metric='Arc')
```

```python
index_stops=tree.query_ball_point(X,  0.15)
```

```python
stop_ids=list(df_stops.iloc[index_stops].stop_id.values)
int(stop_ids[1])
```
