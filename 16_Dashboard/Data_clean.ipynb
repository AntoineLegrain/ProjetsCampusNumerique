```python
import pandas as pd

df_1 = pd.read_csv('./Data/Morticd10_part1.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('./Data/Morticd10_part2.csv')
df_2.head()
```

```python
df = pd.concat([df_1, df_2]).drop_duplicates(keep=False)
df.head()
```

```python
df.tail()
```

```python
df_country = pd.read_csv('./Data/country_codes.csv')
df_country.head()
```

```python
country_dict = df_country.set_index('country').to_dict()
df['Country'] = df['Country'].apply(lambda x: country_dict['name'][x])
```

```python
df.head()
```

```python
df_disease = pd.read_csv('./Data/disease_list.csv', names=['code','name'])
df_disease.head()
```

```python
import numpy as np
main_list = np.setdiff1d(df.Cause.unique(),df_disease.code.unique().astype(str))
```

```python
df_c = df[~df.Cause.isin(main_list)]
df_c.head()
```

```python
df_c['Cause']=df_c.Cause.astype(int)
```

```python
df_c.Cause
```

```python
disease_dict = df_disease.set_index('code').to_dict()
df_c['Cause'] = df_c['Cause'].apply(lambda x: disease_dict['name'][x])
```
