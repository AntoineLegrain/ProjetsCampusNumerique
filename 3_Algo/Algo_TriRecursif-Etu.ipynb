````python
import random
import time
from matplotlib import pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from sorting import *
%load_ext line_profiler

def u_array(debug=False, N=21):
    """
    Retourne un tableau non trie avec des valeur de 1 <= i < N .
    Par defaut un tableau  1 <= i < 21
    
    Pour tester vos algorithmes, vous pouvez passer l'argument debug=True
    dans ce cas, l'algorithme vous donne toujour le meme tableau :
    
    [3, 9, 7, 1, 6, 2, 8, 4, 5]
    
    """
    if debug:
        return [3, 9, 7, 1, 6, 2, 8, 4, 5, 10]
    
    array = list(range(1,N))
    random.shuffle(array)

    return array
````

````python
def quick_sort(deck):
    """
    l'intelligence se trouve au niveau de la partition du tableau
    """
    quick_sort_r(deck, 0, len(deck)-1)

def quick_sort_r(deck, first, last):
    if first<last:
        pivot=partition(deck,first,last)
        quick_sort_r(deck,first,pivot-1)
        quick_sort_r(deck,pivot+1,last)
        
def partition(deck, first, last):
    pivot=deck[first]
    i=first
    j=last
    while i<=j:
        if deck[i]<=pivot:
            i+=1
        else:
            if deck[j]>pivot:
                j-=1
            else:
                swap(deck,i,j)
    swap(deck,first,j)
    return j
````

````python
def merge_sort(deck):
    """
    l'intelligence se trouve dans la fusion des tableaux
    """
    merge_sort_r(deck, 0, len(deck))

def merge_sort_r(deck, start, end):
    if start<end-1:
        middle=int((start+end)/2)
        merge_sort_r(deck,start,middle)
        merge_sort_r(deck,middle,end)
        merge(deck,start,middle,end)
        

def merge(deck, start, middle, end):
    i=start
    j=middle
    temp=[0]*len(deck)
    k=0
    for k in range(start,end):
        if i<middle and j<end:
            if deck[i]<=deck[j]:
                temp[k]=deck[i]
                i+=1
            else:
                temp[k]=deck[j]
                j+=1
        else:
            if i<middle:
                temp[k]=deck[i]
                i+=1
            else:
                temp[k]=deck[j]
                j+=1
    for k in range(start,end):
        deck[k] = temp[k]
````

## Comparaison de performances

````python
values = np.round(np.linspace(10, 10000, 100))
quick_sort_time=[]
merge_sort_time=[]
for i in values:
    deck=u_array(N=int(i))
    
    deck2=deck.copy()
    start=time.time()
    quick_sort(deck2)
    end=time.time()
    quick_sort_time.append(end-start)

    deck2=deck.copy()
    start=time.time()
    merge_sort(deck2)
    end=time.time()
    merge_sort_time.append(end-start)
````

````python
# insertion_time=[3.337860107421875e-06,
#  2.09808349609375e-05,
#  0.0012593269348144531,
#  0.0021779537200927734,
#  0.016594886779785156,
#  0.07392168045043945,
#  0.3154582977294922,
#  2.032505989074707,
#  2.917597532272339,
#  8.08118462562561]

fig = plt.figure()
ax = plt.axes()
plt.title('Running time')
plt.xlabel('Number of cards')
plt.ylabel('Time (s)')
plt.plot(values,quick_sort_time,label='Quick sort')
# plt.plot(values,insertion_time,label='Insertion')
plt.plot(values,merge_sort_time,label='Merge_sort')
plt.legend()
plt.show()
````

````python
fig = plt.figure()
ax = plt.axes()
plt.title('Running time')
plt.xlabel('Number of cards')
plt.ylabel('Time (s)')

plt.plot(values,np.array(quick_sort_time)/np.array(values),label='Quick sort')
plt.plot(values,np.array(merge_sort_time)/np.array(values),label='Merge_sort')


def func(x, a, b):
    return a * np.log2(x) + b

popt, pcov = curve_fit(func,np.array(values),np.array(quick_sort_time)/np.array(values))
print(popt)
plt.plot(values, func(values, *popt), 'r-',label='fit: a=%.3e, b=%.3e' % tuple(popt))

popt, pcov = curve_fit(func,np.array(values),np.array(merge_sort_time)/np.array(values))
print(popt)
plt.plot(values, func(values, *popt), 'r-',label='fit: a=%.3e, b=%.3e' % tuple(popt))

x=np.array(values)
y=np.array(quick_sort_time)/np.array(values)
coefficients = np.polyfit(np.log(x),y,1) # Use log(x) as the input to polyfit.
fit = np.poly1d(coefficients) 
plt.plot(x,fit(np.log(x)),"--", label="fit")

plt.legend()
plt.show()
````

````python
import numpy as np
import time
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
l_listLength = np.linspace(10, 1000, 20)
l_time_quickSort = []
for length in l_listLength:
    l_testList = np.arange(int(length))
    tmpTime = 0
    for i in range(20):
        np.random.shuffle(l_testList)
        t0 = time.time()
        quick_sort(l_testList)
        t1 = time.time()
        tmpTime += (t1 - t0)
    l_time_quickSort.append(tmpTime/20)
plt.plot(l_listLength, l_time_quickSort/l_listLength, '*-', label='quick sort')
# Fit
def func(x, a, b):
    return a * np.log2(x) + b
popt, pcov = curve_fit(func, l_listLength, l_time_quickSort/l_listLength)
print(popt)
plt.plot(l_listLength, func(l_listLength, *popt), 'r-',label='fit: a=%.3e, b=%.3e' % tuple(popt))
plt.legend()
plt.show()
````
