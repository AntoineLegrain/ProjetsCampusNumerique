## Pipeline: fit

After discussing with the team we think that coordinates, development stage and sector address are the more relevant features for now. Note that the selecter can be easily used to add new features to test.<br>
<br>
For now we fill Nas with most frequent values (i.e. using SimpleImputer).

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, TransformerMixin

#Custom columns selecter
class SelectColumns(BaseEstimator, TransformerMixin):
    def __init__(self,list_columns=[]): # no *args or **kargs
        self.list_columns=list_columns
    def fit(self,X,Y=None):
        return self
    def transform(self,X):
        return pd.DataFrame(X, columns=self.list_columns)

#Defining different attributes
all_attribs = ['SOUS_CATEGORIE_DESC','STADEDEDEVELOPPEMENT','LON','LAT','ADR_SECTEUR']
num_attribs = ['LON','LAT']
cat_attribs = ['SOUS_CATEGORIE_DESC','STADEDEDEVELOPPEMENT','ADR_SECTEUR']

#Defining text pipeline: columns selecter, changing Nans by most frequent values, Encoding text values, standardizing
text_pipeline = Pipeline([
('selecter',SelectColumns(list_columns=all_attribs)),
('imputer',SimpleImputer(strategy="most_frequent")) ,
('cat_encoder',OrdinalEncoder()),
 ('std_scaler', StandardScaler())
])

#Defining num pipeline: columns selecter, changing Nans by most frequent values, standardizing
num_pipeline = Pipeline([
('selecter',SelectColumns(list_columns=all_attribs)),
('imputer',SimpleImputer(strategy="most_frequent")),
('std_scaler', StandardScaler())
])

#Full pipeline!!!
full_pipeline = ColumnTransformer([
("cat", text_pipeline, cat_attribs),
("num", num_pipeline, num_attribs)
])

# df_trees_train_na_tr = NA_pipeline.transform(df_trees_train)
# df_trees_train_text_tr = text_pipeline.fit_transform(df_trees_train[cat_attribs])
# df_trees_train_num_tr = num_pipeline.fit_transform(df_trees_train[num_attribs])
# columns_dropping = DropColumns(list_columns=all_attribs)
# df_trees_funky = columns_dropping.transform(df_trees_train)
```

```python
df_trees_train_prepared=pd.DataFrame(full_pipeline.fit_transform(df_trees_train),columns=all_attribs)
df_trees_train_prepared.head(2)
```

## Pipeline: prepare test (transform)

```python
df_trees_test_prepared=pd.DataFrame(full_pipeline.transform(df_trees_test),columns=all_attribs)
df_trees_test_prepared.head(2)
```

# First glance at models

## Linear regression

### Train set

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

lin_reg = LinearRegression()

cv_scores = cross_val_score(lin_reg, df_trees_train_prepared, df_trees_train_Y, cv = 10, scoring='neg_mean_squared_error')

#'neg_mean_squared_error' since the way they actually optimize it 
#is by taking its negative and maximizing that instead of minimizing the actual error.
#Returns the 'negative MSE' for each splits.

#cv is the number of splits (5 by default).

print('Train set')
print('mean MSE:',np.mean(-cv_scores))
print('std MSE:',np.std(-cv_scores))

print('mean RMSE',np.sqrt(np.mean(-cv_scores)))
```

Bad model! 15 years error in average, for a population that is 32 years old in average.

```python
df_trees_train_Y_predict = cross_val_predict(lin_reg, df_trees_train_prepared,df_trees_train_Y, cv = 10)
plt.scatter(df_trees_train_Y,df_trees_train_Y_predict)
```

### Test set

```python
model=LinearRegression()
model.fit(df_trees_train_prepared,df_trees_train_Y)
df_trees_test_Y_predict_lr = model.predict(df_trees_test_prepared)
plt.scatter(df_trees_test_Y,df_trees_test_Y_predict_lr)
```

```python
print('RMSE on prediction on test:')
np.sqrt(mean_squared_error(df_trees_test_Y,df_trees_test_Y_predict_lr))
```

## Elastic net

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV

elastic = ElasticNet(max_iter=10000)

param_grid = [
 {'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]},
 ]

grid_search_elastic = GridSearchCV(elastic, param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search_elastic.fit(df_trees_train_prepared, df_trees_train_Y)

grid_results_elastic = pd.DataFrame(grid_search_elastic.cv_results_)
grid_results_elastic.sort_values(by=['rank_test_score'],inplace=True)
grid_results_elastic
```

```python
df_trees_test_Y_predict_en=grid_search_elastic.best_estimator_.predict(df_trees_test_prepared)
mse_stan_test = mean_squared_error(df_trees_test_Y_predict_en,df_trees_test_Y)
print("test MSE:",mse_stan_test)
print("test RMSE:",np.sqrt(mse_stan_test))
print(30*'-')
```

Elastic net is useless: having a small number of predictors, alpha is nearly equal to 0, being equivalent to least sqaures emthods (classic LR).

# Fine tuning models

We choose to fine tune random forest. The hyperparameter are n_estimators and random_state (is it?).

Best hyperparameter is aroud 250.

```python
rf = RandomForestRegressor(random_state = True)

from pprint import pprint# Look at parameters used by our current forest
print('Parameters currently in use:\n')
pprint(rf.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV# Number of trees in random forest

n_estimators = [int(x) for x in np.linspace(start = 10, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)# Fit the random search model
rf_random.fit(df_trees_train_prepared, df_trees_train_Y)
```

```python
randomized_results = pd.DataFrame(rf_random.cv_results_)
randomized_results.sort_values(by=['rank_test_score'],inplace=True)
randomized_results.head()
```

```python
df_trees_test_Y_predict_rf=rf_random.best_estimator_.predict(df_trees_test_prepared)
rf_random.best_params_
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(df_trees_test_Y, df_trees_test_Y_predict_rf))
print('Mean Squared Error:', metrics.mean_squared_error(df_trees_test_Y, df_trees_test_Y_predict_rf))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(df_trees_test_Y, df_trees_test_Y_predict_rf)))
```

```python
plt.scatter(df_trees_test_Y,df_trees_test_Y_predict_rf)
```
